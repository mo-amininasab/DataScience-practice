## Vector Addition

```python
#جمع به روش متوازی الاضلاع
v1 = np.array([3, 2])
v2 = np.array([0.5,4])
v3 = v1+v2
v3 = np.add(v1,v2)
print('V3 =' ,v3)
plt.plot([0,v1[0]] , [0,v1[1]] , 'r' , label = 'v1')
plt.plot([0,v2[0]] , [0,v2[1]], 'b' , label = 'v2')
plt.plot([0,v3[0]] , [0,v3[1]] , 'g' , label = 'v3')
plt.plot([-2, 8] , [0,0] , 'k--')
plt.plot([0,0] , [-2, 8] , 'k--')
plt.grid()
plt.axis((-2, 8, -2, 8))
plt.legend()
plt.show()
```

```python
#جمع به روش مثلث
plt.plot([0,v1[0]] , [0,v1[1]] , 'r' , label = 'v1')
plt.plot([0,v2[0]] + v1[0] , [0,v2[1]] + v1[1], 'b' , label = 'v2')
plt.plot([0,v3[0]] , [0,v3[1]] , 'g' , label = 'v3')
plt.plot([-2, 8] , [0,0] , 'k--')
plt.plot([0,0] , [-2, 8] , 'k--')
plt.grid()
plt.axis((-2, 8, -2, 8))
plt.legend()
plt.show()
```

```python
#تفاضل دو بردار
v3 = v2 - v1
print('V3 =' ,v3)
plt.plot([0,v1[0]] , [0,v1[1]] , 'r' , label = 'v1')
plt.plot([0,v2[0]] , [0,v2[1]], 'b' , label = 'v2')
plt.plot([v1[0],v2[0]] , [v1[1],v2[1]] , 'g' , label = 'v3')
plt.plot([0,v3[0]] , [0,v3[1]], 'g' , label = 'v2')
plt.plot([-4, 8] , [0,0] , 'k--')
plt.plot([0,0] , [-4, 8] , 'k--')
plt.grid()
plt.axis((-4, 8, -4, 8))
plt.legend()
plt.show()
```

## Scalar Multiplication

```python
v1 = np.array([1, 3])
a, b, c = 0.5, 1.5, -1
v2, v3, v4 = (v1*a), (v1*b), (v1*c)
plt.plot([0,v1[0]] , [0,v1[1]] , 'r' , label = 'v1')
plt.plot([0,v2[0]] , [0,v2[1]], 'b--' , label = 'v2')
plt.plot([0,v3[0]] , [0,v3[1]], 'g:' , label = 'v3')
plt.plot([0,v4[0]] , [0,v4[1]], 'y--' , label = 'v4')

plt.plot([-4,6] , [0,0] , 'k--')
plt.plot([0,0] , [-4,6] , 'k--')
plt.grid()
plt.axis((-4, 6, -4, 6))
plt.legend()
plt.show()
```

```python
u = np.array([1, 3, 2])
v = np.array([2, -1, 0])
#ضرب المان به المان elementwise 
w = np.multiply(u, v)
print(w)
#ضرب داخلی
dotp = u@v
print(" Dot product - ",dotp)
dotp = np.dot(u,v)
print(" Dot product usign np.dot",dotp)
dotp = np.inner(u,v)
print(" Dot product usign np.inner", dotp)
dotp = sum(np.multiply(u,v))
print(" Dot product usign np.multiply & sum",dotp)
dotp = np.matmul(u,v)
print(" Dot product usign np.matmul",dotp)
dotp = 0
for i in range(len(u)):
    dotp = dotp + u[i]*v[i]
print(" Dot product usign for loop" , dotp)
```

## Length of Vector

```python
u = np.array([1, -2, 2])
length = np.sqrt(np.dot(u, u))
print(length)
```

## Normalized Vector

```python
v = np.array([1, -1, np.sqrt(2)])
print(v)
v_len = np.sqrt(np.dot(v, v))
print(v / v_len)
```

## Angle between vectors

```python
v1 = np.array([3, -2])
v2 = np.array([4, 6])
ang = np.rad2deg(np.arccos( np.dot(v1,v2) / (np.linalg.norm(v1)*np.linalg.norm(v2))))
plt.plot([0,v1[0]] , [0,v1[1]] , 'r' , label = 'v1')
plt.plot([0,v2[0]]+v1[0] , [0,v2[1]]+v1[1], 'b' , label = 'v2')
plt.plot([-10,-10] , [0,0] , '--')
plt.plot([0,0] , [-10,10] , '--')
plt.grid()
plt.axis((-10, 10, -10, 10))
plt.legend()
plt.title('Angle between Vectors - %s'  %ang)
plt.show()
```

```python
# 3d vectors
v1 = np.array([1,2,-1])
v2 = np.array([-2,-2,0])
fig = plt.figure()
ax = Axes3D(fig)
ax.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'b')
ax.plot([0, v2[0]],[0, v2[1]],[0, v2[2]],'r')
ang = np.rad2deg(np.arccos( np.dot(v1,v2) / (np.linalg.norm(v1)*np.linalg.norm(v2)) ))
plt.title('Angle between vectors: %s degrees.' %ang)
```
