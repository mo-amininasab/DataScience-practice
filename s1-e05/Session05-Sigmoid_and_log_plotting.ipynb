```python
# sigmoid function
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
print(sigmoid(0))
print(sigmoid(np.inf))
print(sigmoid(-np.inf))

print(sigmoid(0.5))
```

```python
x = np.arange(-4, 12, 0.01)
y = sigmoid(x)
plt.plot(x, y, lw=3)
```

## What do $-log(h_w(x))$ and $-log(1-h_w(x))$ look like?

```python
x = np.arange(0.01, 1, 0.01)
```

```python
plt.figure(figsize=(12, 4), dpi=150)
plt.subplot(1, 3, 1)
plt.plot(x, np.log(x))
plt.title('log(x)')

plt.subplot(1, 3, 2)
plt.plot(x, -np.log(x))
plt.title('-log(x)')

plt.subplot(1, 3, 3)
plt.plot(x, -np.log(1-x))
plt.title('-log(1-x)')
```
