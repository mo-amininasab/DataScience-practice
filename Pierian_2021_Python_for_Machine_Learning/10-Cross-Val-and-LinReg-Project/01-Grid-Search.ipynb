# Grid Search

We can search through a variety of combinations of hyperparameters with a grid search. While many linear models are quite simple and even come with their own specialized versions that do a search for you, this method of a grid search will can be applied to _any_ model from sklearn, and we will need to use it later on for more complex models, such as Support Vector Machines.


## Imports


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data


```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

### Formatting Data


```python
## CREATE X and y
X = df.drop('sales', axis=1)
y = df['sales']

# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=101)

# SCALE DATA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

## Model


```python
from sklearn.linear_model import ElasticNet
```

```python
# help(ElasticNet)
```

```python
base_elastic_model = ElasticNet()
```

## Grid Search

A search consists of:

- an estimator (regressor or classifier such as sklearn.svm.SVC());
- a parameter space;
- a method for searching or sampling candidates;
- a cross-validation scheme
- a score function.


```python
# dictionary of parameters to search
param_grid = {
    'alpha': [0.1, 1, 5, 10, 50, 100],
    'l1_ratio': [.1, .5, .7, .9, .95, .99, 1]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# verbose number a personal preference
grid_model = GridSearchCV(estimator=base_elastic_model,
                          param_grid=param_grid,
                          scoring='neg_mean_squared_error',
                          cv=5,
                          verbose=2)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
# pd.DataFrame(grid_model.cv_results_)
```

## Using Best Model From Grid Search


```python
y_pred = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)
```
