## Cropping

By left-top and right-bottom vertex.


```python
left = 50
top = 100
right = 300
bottom = 400
cropped = im.crop((left, top, right, bottom))
plt.imshow(cropped)
```

## Transposing


```python
im_transposed_01 = im.transpose(Image.ROTATE_180)
im_transposed_02 = im.transpose(Image.FLIP_LEFT_RIGHT)
im_transposed_03 = im.transpose(Image.FLIP_TOP_BOTTOM)
im_transposed_04 = im.transpose(Image.ROTATE_90)
im_transposed_05 = im.transpose(Image.ROTATE_180)
im_transposed_06 = im.transpose(Image.ROTATE_270)

# plt.figure(figsize=(10,10))

plt.subplot(321)
plt.imshow(im_transposed_01)

plt.subplot(322)
plt.imshow(im_transposed_02)

plt.subplot(323)
plt.imshow(im_transposed_03)

plt.subplot(324)
plt.imshow(im_transposed_04)

plt.subplot(325)
plt.imshow(im_transposed_05)

plt.subplot(326)
plt.imshow(im_transposed_06)
```

## Resizing


```python
new_size = (50, 50)
out = im.resize(new_size)
plt.imshow(out)
```

### Interpolation Techniques

1. BILINEAR
2. NEAREST
3. BOX
4. Hamming
5. BICUBIC
6. LANCZOS


```python
im_resized_01 = im.resize(new_size, Image.BILINEAR)
im_resized_02 = im.resize(new_size, Image.NEAREST)
im_resized_03 = im.resize(new_size, Image.BOX)
im_resized_04 = im.resize(new_size, Image.HAMMING)
im_resized_05 = im.resize(new_size, Image.BICUBIC)
im_resized_06 = im.resize(new_size, Image.LANCZOS)  # Better Performance

plt.figure(figsize=(10, 10))

plt.subplot(321)
plt.imshow(im_resized_01)

plt.subplot(322)
plt.imshow(im_resized_02)

plt.subplot(323)
plt.imshow(im_resized_03)

plt.subplot(324)
plt.imshow(im_resized_04)

plt.subplot(325)
plt.imshow(im_resized_05)

plt.subplot(326)
plt.imshow(im_resized_06)
```

## Rotating

## Text Watermark


```python
from PIL import ImageFont
from PIL import ImageDraw

text_marked = ImageDraw.Draw(im_copy)
#font = ImageFont.load_default()
font = ImageFont.truetype('arial.ttf', 100)  # choose your favorite
text_marked.text((100, 100), "It's a test", (0, 0, 0), font=font)
plt.imshow(im_copy)
```

## Image Enahancement


```python
from PIL import ImageEnhance

im_enhanced_01 = ImageEnhance.Color(im).enhance(3)
im_enhanced_02 = ImageEnhance.Contrast(im).enhance(3)
im_enhanced_03 = ImageEnhance.Brightness(im).enhance(3)
im_enhanced_04 = ImageEnhance.Sharpness(im).enhance(3)

plt.figure(figsize=(10, 10))

plt.subplot(221)
plt.imshow(im_enhanced_01)

plt.subplot(222)
plt.imshow(im_enhanced_02)

plt.subplot(223)
plt.imshow(im_enhanced_03)

plt.subplot(224)
plt.imshow(im_enhanced_04)
```
