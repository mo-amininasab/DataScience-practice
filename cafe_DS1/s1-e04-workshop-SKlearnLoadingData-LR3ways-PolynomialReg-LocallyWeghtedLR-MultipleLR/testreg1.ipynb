```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor, LinearRegression
```

```python
dataset = pd.read_csv('Salary_data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
dataset.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

## Trying SGD

```python
regressor = SGDRegressor()
regressor.fit(X_train, y_train)
```

```python
w0_SGD = regressor.intercept_
w1_SGD = regressor.coef_
print('w0_SGD: ', w0_SGD)
print('w1_SGD: ', w1_SGD)
```

```python
regressor.score(X_test, y_test)
```

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue');

plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Trying NE

```python
regressor2 = LinearRegression()
regressor2.fit(X_train, y_train)
```

```python
w0_LR = regressor2.intercept_
w1_LR = regressor2.coef_
print('w0_LR: ', w0_LR)
print('w1_LR: ', w1_LR)
```

```python
regressor2.score(X_test, y_test)
```

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor2.predict(X_train), color='blue');

plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
print('15 years of experience: ', regressor2.predict([[15]]))
print('20 years of experience: ', regressor2.predict([[20]]))

print('15 years of experience: ', regressor.predict([[15]]))
print('20 years of experience: ', regressor.predict([[20]]))
```
